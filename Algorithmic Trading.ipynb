### Custom Function

```python
def plot_moving_averages(df, stock_name, figsize=(20, 10), title='Stock Price and Moving Averages'):
    """
    Plots stock price, 50-day moving average, and 200-day moving average.

    Parameters:
    - df: DataFrame containing columns 'Close', 'SMA50', and 'SMA200'.
    - stock_name: String, the name of the stock.
    - figsize: Tuple, optional, default is (20, 10). Size of the figure.
    - title: String, optional, default is 'Stock Price and Moving Averages'. Title of the plot.

    Returns:
    None
    """
    # Set Seaborn style
    sns.set(style="whitegrid")

    # Plotting stock price, 50-day, and 200-day moving averages with improved styling
    plt.figure(figsize=figsize)

    # Plotting stock price
    plt.plot(df.index, df['Close'], label='Stock Price', color='b', linewidth=2)

    # Plotting 50-day moving average
    plt.plot(df.index, df['SMA50'], label='50-day MA', color='g', linestyle='--', linewidth=2)

    # Plotting 200-day moving average
    plt.plot(df.index, df['SMA200'], label='200-day MA', color='r', linestyle='-.', linewidth=2)

    # Adding title with stock name and labels
    plt.title(f"{stock_name} {title}", fontsize=18)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Price', fontsize=14)

    # Adding a grid for better readability
    plt.grid(True, linestyle='--', alpha=0.7)

    # Adding legend with larger font size
    plt.legend(loc='upper left', fontsize=14)

    # Show the plot
    plt.show()
```

## Graphing 5 stocks

```python
plot_moving_averages(nvda, 'NVDIA')
plot_moving_averages(fico, 'FICO')
plot_moving_averages(cost, 'Costco')
plot_moving_averages(aapl, 'Apple')
plot_moving_averages(ma, 'MasterCard')
```

## SMA Strategy

```python
data = apple.Close.loc[(apple.index>='1991-01-01')]
```

```python
data = data.to_frame()
```

```python
data
```

```python
# moving averages
sma_s = 50
sma_l = 100

data['sma_s'] = data.Close.rolling(sma_s).mean()
data['sma_l'] = data.Close.rolling(sma_l).mean()
```

```python
data
```

```python
data.plot(figsize=(12,8), title='Apple - SMA {} | SMA {}'.format(sma_s, sma_l), fontsize=12);
```

```python
data.loc['2021'].plot(figsize=(12,8), title='Apple - SMA {} | SMA {}'.format(sma_s, sma_l), fontsize=12);
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['position'] = np.where(data['sma_s']>data['sma_l'], 1, -1)
```

```python
data
```

```python
data.loc['2023',["sma_s", "sma_l",'position']].plot(figsize=(12,8), title='Apple - SMA {} | SMA {}'.format(sma_s, sma_l), fontsize=12,
                                              secondary_y='position');
```

```python
# buy and hold column
data['returns_bnh'] = np.log(data.Close.div(data.Close.shift(1)))
```

```python
data
```

```python
# strategy column
data['strategy'] = data['returns_bnh'] * data.position.shift(1)
```
