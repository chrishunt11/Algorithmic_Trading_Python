## Exploring AAPL stock

```python
aapl = close.AAPL.copy().to_frame() # convert data to DataFrame
```

```python
aapl
```

```python
aapl.shift(periods=1)
```

```python
aapl["lag1"] = aapl.shift(periods=1)
```

```python
aapl
```

```python
# calculate the difference
aapl['Diff'] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl
```

```python
aapl['% Change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl
```

```python
# Using the diff method
aapl['Diff2'] = aapl.AAPL.diff(periods=1)
aapl
```

```python
aapl['% Change 2'] = aapl.AAPL.pct_change(periods=1).mul(100)
aapl
```

```python
aapl
```

```python
aapl.rename(columns={'% Change 2': '% Change'}, inplace=True)
```

```python
aapl
```

### Resample

```python
# Getting all of the data at the end of the month
aapl.AAPL.resample("M").last()
```

```python
# last business day of the month
aapl.AAPL.resample("BM").last()
```

```python
# last business day of the month percent change
aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

```python
del aapl['% Change']
del aapl['lag1']
del aapl['Diff']
del aapl['Diff2']
```

```python
ticker = ["SPY", "KO", "AAPL", "MSFT", "IBM", "DIS"]
```

```python
stocks = yf.download(ticker, start="2010-01-01", end="2023-11-30")
```

```python
stocks
```

```python
close = stocks.loc[:,"Close"].copy()
close
```

```python
normal_close = close.div(close.iloc[0]).mul(100)
```

```python
normal_close = close.div(close.iloc[0]).mul(100)
normal_close.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe().T
```

```python
summary = ret.describe().T.loc[:,["mean", "std"]]
summary
```

```python
summary['mean'] = summary['mean'] * 252
summary['std'] = summary['std'] * np.sqrt(252)
summary
```

```python
summary.plot.scatter(x='std', y='mean', figsize=(12, 8), s=50, fontsize=15, c='blue', marker='o')
plt.grid(True)

for i in summary.index:
    plt.annotate(f"{i}\nRisk: {summary.loc[i,'std']:.4f}\nReturn: {summary.loc[i,'mean']:.4f}", 
                 xy=(summary.loc[i,'std']+0.002, summary.loc[i,'mean']+0.002), 
                 size=12, ha='left', va='bottom')

plt.xlabel("Annual Risk (std)", fontsize=15)
plt.ylabel("Annual Return (mean)", fontsize=15)
plt.title("Risk/Return", fontsize=20);
```
